Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd
import seaborn as sns
import json
```

```python
df1 = pd.read_csv('cb_data_frame.csv')
df2 = pd.read_csv('yelp_data_frame.csv')
df3 = pd.read_csv('fsq_data_frame.csv')
```

```python
df1['category'] = 'city_bike'
df2['category'] = 'yelp_restaruant'
df3['category'] = 'fsq_restaruant'
```

```python
joined_df = pd.merge(df3, df2, on=['name', 'latitude', 'longitude', 'category'],  how='outer')
```

```python
new_merged_df = pd.merge(joined_df, df1, on=['name', 'latitude', 'longitude', 'category'], how='outer')
```
